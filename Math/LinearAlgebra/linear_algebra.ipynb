## Vector

A vector is a quantity that has both magnitude and direction. It is commonly represented as an array of numbers that define its components.
A vector is an ordered list of numbers, often written as: $v = [v1, v2,... v3]$

A Two Dimensional Vector (2D) has $x$ and $y$ components and Three Dimensional (3D) vector have $x$, $y$, $z$ components and dimension can get complicated it we working with multiple vectors and multiple components

A 2D vector might look like  $v=[3,4]$
A 3D vector might look like $v=[1,−2,5]$

### Geometric Vectors

A vector can be thought of as a directed line segment in space.
In 2D, the vector $v=[3,4]$ can be visualized as an arrow starting at the origin (0,0) and ending at the point (3,4). like $x-axis$ and $y-axis$
and It plays important role in game development and it and vector ca extend to multiple dimensions such as multiple axises


### Polynomial Vectors

A polynomial vector is a representation of a polynomial as a row vector of coefficients. For example, the polynomial $x^{3}+2x^{2}-4x+3$ can be represented as the vector $[1,2,-4,3]$. it's like value of each vector

### Operations of Vectors

- #### Addition:
Two vectors can be added together and multiple vectors can also be added to each other and resulting the another vector

$x = [x1, x2,....xn]$

$y = [y1,y2,......yn]$

$x + y = [x1 + y1, x2 + y2,.....xn + yn]$

- #### Subtraction:
Two vectors can be subtracted together and multiple vectors can also be subtracted from each other and resulting the another vector

$x = [x1, x2,....xn]$

$y = [y1,y2,......yn]$

$x - y = [x1 - y1, x2 - y2,.....xn - yn]$

- #### Multiplication (Dot Product):

Multiplication of two or more vectors result another vector while Dot Product result and single value a real number (Scalar)
In dot product first they all multiplied together and added but in multiplication they all just multiplied.


- #### Magnitude of a Vector
To calculate the magnitude of a vector, you can use the formula $|v|=√{x^{2}+y^{2}}$ for a two-dimensional vector, or $|V|=√{x^{2}+y^{2}+z^{2}}$ for a three-dimensional vector

The magnitude of a vector is its length.

Magnitude: The formula $\|\vec{v}\|=√v_{1}^{2}+v_{2}^{2}+\cdots +v_{n}^{2}$ is used to calculate the magnitude of a vector $vec{v}$ in $n$-dimensional space. 


### Types of Vectors

There are 10 types of vectors in mathematics which are

- #### Zero Vector
A zero vector is a vector when the magnitude of the vector is zero and the starting point of the vector coincides with the terminal point.
i.e $v = [0,0,0]$ in this example of the vector all $x, y, z,$ coordinates are zero so it's a zero vector

- #### Unit Vector
A vector with a magnitude of one. i.e if the length of the vector is 1 is unit vector
- #### Position vector
It shows the position of the vector from origin to it's ending point i.e $v = [1, 5]$ such as the vector start from position 1 to the ending position 5.

- #### Equal vector
Two vectors that have the same magnitude and direction, regardless of where their starting points are
- #### Negative of a vector
A vector that has the same magnitude as another vector, but its direction is opposite



```python
import numpy as np

## Vectors have to equal in sizes

vectorOne = np.array([1,1,1])
vectorTwo = np.array([2,2,2] )

# Addition

print("Addition:" ,vectorOne + vectorTwo)

# Subtraction

print("Subtraction:" ,vectorOne - vectorTwo)

# Dot Product || Multiplication

print("Multiplication:", vectorOne * vectorTwo )
print("Dot Product:", np.dot(vectorOne, vectorTwo))

# Cross Product (Only work for more than two vectors i,e 3Dimensional)

print("Cross Product:", np.cross(vectorOne, vectorTwo))

# Magnitude of vectors

print("VectorOne Magnitude:", np.linalg.norm(vectorOne))
print("VectorTwo Magnitude:", np.linalg.norm(vectorTwo))

# Normalization of Vector (Unit Vector)

print("vectorOneMagnitude:", vectorOne / np.linalg.norm(vectorOne))
print("vectorTwoMagnitude:", vectorTwo / np.linalg.norm(vectorTwo))

```

## System of Linear Equations
A system of linear equations (or linear system) is a collection of two or more linear equations involving the same variables.

 - The system of one equation in one unknown 

    $2x = 4$

    $x = 4/2$

    $x = 2$

- The simplest kind of nontrivial linear system involves two equations and two variables:

    $2x + 3y = 6$

    $4x + 9y = 15$

    first we solve the first equation with respect to x and when we find the value of x we put it into the second equation and find the value of y and than by putting each x and  y values in both equations we solve the system of linear equations in traditional manners
